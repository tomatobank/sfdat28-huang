The Health Department has developed an inspection report and scoring system. 
After conducting an inspection of the facility, the Health Inspector calculates a score based on the violations observed. 
Violations can fall into:
- **high risk category**: records specific violations that directly relate to the transmission of food borne illnesses,the adulteration of food products and the contamination of food-contact surfaces.
- **moderate risk category**: records specific violations that are of a moderate risk to the public health and safety.
- **low risk category**: records violations that are low risk or have no immediate risk to the public health and safety.

```python
businesses = pd.read_csv('./data/businesses_plus.csv', parse_dates=True, dtype={'phone_number': str})
businesses.head()
# dtype casts the column as a specific data type
```

```python
inspections = pd.read_csv('./data/inspections_plus.csv', parse_dates=True)
inspections.head()
```

```python
violations = pd.read_csv('./data/violations_plus.csv', parse_dates=True)
violations.head()
```

```python
# 1 Combine the three dataframes into one data frame called restaurant_scores
# Hint: http://pandas.pydata.org/pandas-docs/stable/merging.html

restaurant_scores = businesses.merge(inspections,on='business_id').merge(violations,on=['business_id','date'])
restaurant_scores.head()
```

```python
# 2 Which ten business have had the most inspections?
restaurant_scores.groupby('business_id').date.nunique().sort_values(ascending=False).head(10)
```

```python
# 3 Group and count the inspections by type
inspections.groupby('type').count()
```

```python
# 4 Create a plot that shows number of inspections per month
# Bonus for creating a heatmap
# http://stanford.edu/~mwaskom/software/seaborn/generated/seaborn.heatmap.html?highlight=heatmap

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

restaurant_scores['date'] = pd.to_datetime(restaurant_scores['date'],format='%Y%m%d')
restaurant_scores.date.dt.month.value_counts().sort_index().plot(kind='bar')
```

```python
# 5 Which zip code contains the most high risk violations?
restaurant_scores.groupby('postal_code').count()['ViolationTypeID'].sort_values(ascending=False)
```

```python
# 6 If inspection is prompted by a change in restaurant ownership, 
# is the inspection more likely to be categorized as higher or lower risk?

restaurant_scores[restaurant_scores['type'] == "New Ownership"].groupby('risk_category').count()
```

```python
# 7 Examining the descriptions, what is the most common violation?
restaurant_scores.description.value_counts()
```

```python
# 8 Create a hist of the scores with 10 bins
restaurant_scores['Score'].hist(bins=10)
```

```python
killings = pd.read_csv('./data/police-killings.csv')
killings.head()
```

```python
# 1. Make the following changed to column names:
# lawenforcementagency -> agency
# raceethnicity        -> race
killings = killings.rename(columns = {'lawenforcementagency':'agency','raceethnicity':'race'})
killings.columns
```

```python
# 2. Show the count of missing values in each column
killings.isnull().sum()
```

```python
# 3. replace each null value in the dataframe with the string "Unknown"
killings.streetaddress.where(killings.streetaddress.isnull(), True,False)
```

```python
# 4. How many killings were there so far in 2015?
killings['year'][killings['year'] == 2015].count()
```

```python
# 5. Of all killings, how many were male and how many female?
```

```python
killings.groupby('gender')['gender'].count()
```

```python
# 6. How many killings were of unarmed people?
killings[killings.armed == 'No']['armed'].count()
```

```python
# 7. What percentage of all killings were unarmed?
killings[killings.armed == 'No']['armed'].count() / float(len(killings))
```

```python
# 8. What are the 5 states with the most killings?
killings.groupby('state')['state'].count().sort_values(ascending=False).head(5)
```

```python
# 9. Show a value counts of deaths for each race
killings.groupby('race')['race'].count()
```

```python
# 10. Display a histogram of ages of all killings
killings.age.hist(bins=10)
```

```python
# 11. Show 6 histograms of ages by race
killings.age.hist(by=killings.race, bins=10, sharex=True, sharey=True)
```

```python
# 12. What is the average age of death by race?
killings.groupby('race').mean()['age']
```

```python
# 13. Show a bar chart with counts of deaths every month
killings.groupby('month')['month'].count().sort_index().plot(kind = 'bar')
```

```python
majors = pd.read_csv('./data/college-majors.csv')
majors.head()
```

```python
# 1. Delete the columns (employed_full_time_year_round, major_code)

majors
```

```python
# 2. Show the count of missing values in each column
majors.isnull().sum()
```

```python
# 3. What are the top 10 highest paying majors?
majors[['Major','P75th']].sort_values(by='P75th',ascending = False).head(10)
```

```python
# 4. Plot the data from the last question in a bar chart, include proper title, and labels!

majors[['Major','P75th']].sort_values(by='P75th',ascending = False).head(10).plot(kind='bar',x='Major',y='P75th')
plt.xlabel("Majorsssss")
plt.ylabel("P75th")
```

```python
majors.groupby('Major')
```

```python
# 5. What is the average median salary for each major category?
majors.groupby('Major_category').mean()['Median'].sort_values(ascending = False)
```

```python
# 6. Show only the top 5 paying major categories
majors.groupby('Major_category').mean()['P75th'].sort_values(ascending = False).head(5)
```

```python
# 7. Plot a histogram of the distribution of median salaries
majors['Median'].hist(bins = 10)
```

```python
# 8. Plot a histogram of the distribution of median salaries by major category
majors['Median'].hist(by=majors.Major_category, bins = 10, sharex=True, sharey=True)
```

```python
# 9. What are the top 10 most UNemployed majors?
majors.groupby('Major')['Unemployed'].sum().sort_values(ascending=False).head(10)
```

```python
# What are the unemployment rates?
majors.groupby('Major')['Unemployed','Unemployment_rate'].sum().sort_values(by='Unemployed', ascending=False).head(10)
```

```python
# 10. What are the top 10 most UNemployed majors CATEGORIES? Use the mean for each category
# What are the unemployment rates?

majors.groupby('Major_category')['Unemployed','Unemployment_rate'].mean().sort_values(by='Unemployment_rate', ascending=False).head(10)
```

```python
# 11. the total and employed column refer to the people that were surveyed.
# Create a new column showing the emlpoyment rate of the people surveyed for each major
# call it "sample_employment_rate"
# Example the first row has total: 128148 and employed: 90245. it's 
# sample_employment_rate should be 90245.0 / 128148.0 = .7042

majors['sample_employment_rate'] = majors.Employed / majors.Total
```

```python
# 12. Create a "sample_unemployment_rate" column
# this column should be 1 - "sample_employment_rate"
majors['sample_unemployment_rate'] = 1 - majors.sample_employment_rate
```
